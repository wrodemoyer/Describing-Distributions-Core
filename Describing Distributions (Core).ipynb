## Imports 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
df = pd.read_csv('https://docs.google.com/spreadsheets/d/1APV3pXiAszS_0mSgkiEt9IUNH-QmyX7KwxSAwuADl6Y/gviz/tq?tqx=out:csv&sheet=medical_data')
df.head()
```

```python
df.info()
```

### VitD_levels column

```python
vdl = 'VitD_levels'
feature= df[vdl]
mean1= df['VitD_levels'].mean()
print(f'Mean: {mean1}')
median1= df['VitD_levels'].median()
print(f'Median: {median1}')
std= feature.std()
plus_one_std= mean1 + std
minus_one_std= mean1 - std
```

```python
fig, ax = plt.subplots(figsize= (10,6))
sns.histplot(feature, ax= ax, stat= 'probability', kde= True)
ax.axvline(mean1, color = 'red',label=f'Mean = {mean1:,.2f}')
ax.axvline(median1, ls='--', color = 'green', label = f'Median = {median1:,.2f}')
ax.axvline(plus_one_std, color = 'black',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color = 'black', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(plus_one_std, minus_one_std, color = 'yellow', zorder = 0)
ax.set_title(f'{vdl}')
ax.legend();
```

```python
kurt = stats.kurtosis(feature, fisher = False)
kurt
```

- Is it Discrete or Continuous?
    - **Continuous**
- Does it have a skew? If so, which direction (+/-)
    - **Normal (No Skew)**
- What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
    - **Mesokurtic**
